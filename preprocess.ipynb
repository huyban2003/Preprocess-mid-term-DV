## 2.4 Kiểm tra số lượng thí sinh thi và số môn thi để tìm thử phương án xử lý

```python
dong_so_mon_thi_bang_0 = raw_df[raw_df[['Ngu Van', 'Toan Hoc', 'Vat Ly', 'Hoa Hoc', 'Sinh Hoc', 'Lich Su', 'Dia Ly', 'GDCD', 'Ngoai Ngu']].count(axis=1) == 0]

# In ra dòng đầu tiên mà số môn thi bằng 0 (nếu có)
if not dong_so_mon_thi_bang_0.empty:
    print("Dòng mà số môn thi bằng 0:")
    print(dong_so_mon_thi_bang_0)
else:
    print("Không có dòng nào mà số môn thi bằng 0.")
```

```python
def dem_so_mon_thi(df):
    # Tạo một Series mới để lưu số lượng môn thi của từng thí sinh
    so_mon_thi = df[['Ngu Van', 'Toan Hoc', 'Vat Ly', 'Hoa Hoc', 'Sinh Hoc', 'Lich Su', 'Dia Ly', 'GDCD', 'Ngoai Ngu']].count(axis=1)
    
    # Đếm số lượng thí sinh thi mỗi số môn
    so_luong_moi_so_mon = so_mon_thi.value_counts().sort_index()
    
    # In ra số lượng môn thi của từng thí sinh
    for so_mon, so_thi_sinh in so_luong_moi_so_mon.items():
        print(f"{so_mon}: {so_thi_sinh}")

# Gọi hàm với dataframe df
dem_so_mon_thi(raw_df)
```

Dựa vào thông tin kiểm tra, để phù hợp cho việc xử lý và nhẹ nhàng cho việc fill dữ liệu trống, ta sẽ loại bỏ các dòng dữ liệu có số môn ` nhỏ hơn 3`

```python
raw_df = raw_df[raw_df[['Ngu Van', 'Toan Hoc', 'Vat Ly', 'Hoa Hoc', 'Sinh Hoc', 
                        'Lich Su', 'Dia Ly', 'GDCD', 'Ngoai Ngu']].count(axis=1) >= 3].reset_index(drop=True)
raw_df
```

```python
dem_so_mon_thi(raw_df)
```
